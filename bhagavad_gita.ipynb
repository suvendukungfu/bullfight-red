ADDING DATA FROM TABLES

IMPORT LIBRARIES

```python
import pandas as pd
import matplotlib.pyplot as plt
```

CREATE DATAFRAMES OF BHAGAVAD GITA

```python
data={'ADHYAY NAME': [
        'Arjuna Vishad Yog', 'Sankhya Yog', 'Karm Yog', 'Gyan Karm Sanyas Yog',
        'Karm Sanyas Yog', 'Dhyan Yog', 'Gyan Vigyan Yog', 'Akshar Brahma Yog',
        'Raj Vidhya Raj Guhya Yog', 'Vibhuti Yog', 'Vishwaroop Darshan Yog',
        'Bhakti Yog', 'Kshetra Kshetragya Vibhag Yog', 'Gunatray Vibhag Yog',
        'Purushottam Yog', 'Daivasur Sampad Vibhag Yog', 'Shraddha Traya Vibhag Yog',
        'Moksha Sanyas Yog'
    ],
    'No_of_Shlokas': [47, 72, 43, 42, 29, 47, 30, 28, 34, 42, 55, 20, 34, 27, 20, 24, 28, 78]
}
df = pd.DataFrame(data)
df
```

SORT AND RANK ADHYAY NAME

```python
df = df.sort_values(by='No_of_Shlokas', ascending=False).reset_index(drop=True)
```

```python
df['Rank'] = df.index + 1
```

```python
df
```

 PLOT Log-Log GRAPH (FOR ZIPFâ€™s LAW)

```python
plt.figure(figsize=(8,5))
plt.loglog(df['Rank'], df['No_of_Shlokas'], marker="o", color='blue')
plt.title('Zipf\'s Law: Bhagavad Gita Chapter Shloka Analysis', fontsize=14)
plt.xlabel('Rank (Log Scale)', fontsize=12)
plt.ylabel('Number of Shlokas (Log Scale)', fontsize=12)
plt.grid(True, which="both", linestyle='--', linewidth=0.5)
plt.show()
```

OBSERVATION

*The plot shows that the number of shlokas in each chapter roughly follows a straight line when plotted on a log-log scale, indicating that Zipfâ€™s law holds approximately. Deviations might be due to thematic importance of certain chapters like "Moksha Sanyas Yog" having more verses.*
